# Expolaritory Data Analysis (EDA):
Expolaratory Data Analysis (EDA) is a critical step in the data analysis process. It involves summarizing the main characteristics of a dataset, often using visual methods. EDA helps to understand the data better, identify patterns, spot anomalies, and check assumptions through statistical graphics and other data visualization methods.
# EDA is important because:
 1. **Understanding the Data**: EDA helps to understand the structure, relationships, and patterns in the data.
 2. **Identifying Anomalies**: It helps to identify outliers or anomalies that may affect the analysis
3. **Checking Assumptions**: EDA allows you to check the assumptions of statistical tests and models.
 4. **Feature Selection**: It helps in selecting the most relevant features for modeling.
 5. **Guiding Further Analysis**: EDA can guide the choice of further analysis techniques and models.
 6. **Communicating Results**: EDA provides visualizations that can help communicate findings to stakeholders.

### Seabron Functionality:
- Seaborn offers a lot of functionality which makes it effective for many data visualization tasks. Some of the key features include:
- Seaborn provides a dataset-oriented API to examine relationships between multiple variables.
- It provides function to fit and visualize linear regression models for different types of independent and dependent variables.
- Seaborn provides functions for visualizing univariate and bivariate distributions for comparing them between different subsets of the data.
- It provides plotting functions for visualizing categorical data.
- It helps us to visualize the matrics of data and use clustering algorithms to discover patterns in the data.
- Seaborn provides a plotting function to plot statistical timeseries data. The function provides flexible estimation and representation of uncertainty around the estimate.
- It provides tool for choosing style ,color palette and context for the plot.
  


### Seaborn Color Palettes:
- Seaborn provides a variety of color palettes that can be used to enhance the aesthetics of your plots. Some common color palettes include:
  - **Deep**: Default color palette with a range of colors.
  - **Muted**: Softer colors that are less saturated.
  - **Pastel**: Light and soft colors.
  - **Dark**: Darker shades of colors.
  - **Colorblind**: Colors that are friendly for colorblind viewers.
  - **Diverging**: Color palettes that diverge from a central point, useful for visualizing differences.
  - **Sequential**: Color palettes that represent a range of values from low to high.
- Seaborn also allows you to create custom color palettes using the `color_palette()` function, where you can specify the colors you want to use in your plots.

### Simple Graphs:

```python
# create a figure and axis
plt.figure(figsize=(10,5))
sns.countplot(x='pclass', hue='survived', data= df)
plt.title('Survived rate by passenger class')
plt.xlabel('Passenger class')
plt.ylabel('count')
plt.legend(labels=['Did not survived', 'survived'])
plt.show()
```

```python
# survivel by gender
plt.figure(figsize=(10,5))
sns.countplot(x='sex', hue='survived', data= df)
plt.title('Survived rate by gender')
plt.xlabel('gender')
plt.ylabel('count')
plt.legend(labels=['Did not survived', 'survived'])
plt.show()
```

```python
# age distribution by age
plt.figure(figsize=(10,5))
sns.histplot(x='age',bins= 30 , kde= True, data= df)
plt.title('Age distribution of passenger')
plt.xlabel('gender')
plt.ylabel('count')
plt.axvline(df['age'].mean(),color= 'red', linestyle= '--',label = 'Mean Age')
plt.legend()
plt.show()
```
