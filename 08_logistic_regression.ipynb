```python
# split the data into x and y 
x = df.drop('Outcome', axis=1)
y = df['Outcome']
```

```python
y.shape
```

## Scaled the data

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

## Split data into training and testing sets

```python
x_train , y_train, x_test, y_test =train_test_split(x_scaled,y, test_size=0.2 ,random_state=42)
```

## Train on Logistic Regression Model

```python
# initialize the model
lr = LogisticRegression()

# Train the model
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=42)
lr.fit(x_train, y_train)
```

## Model Evaluation

```python
# predict the model

y_pred= lr.predict(x_test)
```

```python
# calculate accuracy 
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy : {accuracy:.2f}")


## classification report 
print("\nClassification Report:")
print(classification_report(y_test, y_pred))


# confusion metrix
print("\nConfusion Metrix")
print(confusion_matrix(y_test, y_pred))
```

## Decision Tree Classifier


```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()

model.fit(x_train, y_train)

y_pred = model.predict(x_test)
print("This is decision tree classifier")
print(f"Accuracy Score :{accuracy_score(y_test, y_pred)}")
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()

model.fit(x_train,y_train)

y_pred = model.predict(x_test)
print("This is random forest classifier")
print(f"accuracy Score is :{accuracy_score(y_test, y_pred)}")
```
