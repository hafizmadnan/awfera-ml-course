# Linear Regression
Linear regresssion is a method to find a relationship between a dependent variable
and one or more independent variables using a straight line.
It is a supervised learning algorithm used for regression problems.

#### Mathematics of Linear Regression
The equation of a straight line is given by:

\[ y = mx + b \]
Where:
- \( y \) is the dependent variable (output)        
- \( x \) is the independent variable (input)
- \( m \) is the slope of the line (coefficient)
- \( b \) is the y-intercept (constant term)


## Types of Linear Regression
1.` Simple Linear Regression`:
It is used when there is only one independent variable and one dependent variable.

2.` Multiple Linear Regression`:
It is used when there are multiple independent variables and one dependent variable.

3.` Polynomial Regression`:
It is used when the relationship between the independent and dependent variable is not linear.

4.` Ridge Regression`:
It is used when there is multicollinearity in the data. It adds a penalty term to the loss function to reduce the complexity of the model.

5.` Lasso Regression`:
Another regularization technique that adds a penalty term to the loss function to reduce the complexity of the model. It is used when there are many features in the data and we want to select only a few of them.

#### Evalution Metrics for Linear Regression
1. `Mean Absolute Error (MAE)`:
2. `Mean Squared Error (MSE)`:
3. `Root Mean Squared Error (RMSE)`:
4. `R-squared (R2)`:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv("employee_data_1000.csv")
df.head(3)
```

```python
df.dtypes
```

```python
# split the data into x and y
x = df[['Age']]
# scaler = MinMaxScaler()
# x = scaler.fit_transform(x)
y =df['Salary']
```

```python
print(x.shape)
print(y.shape)
```

```python
# split the data into train and test
x_train , x_test , y_train, y_test = train_test_split(x,y, test_size= 0.2)
```

```python
# call the model
model = LinearRegression()
```

```python
#fit the  model
model.fit(x_train, y_train)
```

```python
# predict the model
y_predict = model.predict(x_test)
```

```python
# display model cofficent 
print("\nModel cofficents:")
print(f"Intercept:{model.intercept_}")
print(f"Slope: {model.coef_[0]}")
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['Age'], df['Salary'])
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Age vs Salary')
plt.show()
```

```python
# PLOT THE MODEL
plt.scatter(x_test, y_test)
plt.plot(x_test, y_predict, color='red')
plt.show()
```
